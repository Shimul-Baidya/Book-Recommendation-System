```python
books_with_ratings = ratings.merge(books, on = 'ISBN')
```

```python
books_with_ratings.head()
```

```python
num_ratings = books_with_ratings.groupby('title')['rating'].count().reset_index()
```

```python
num_ratings.head()
```

```python
num_ratings.rename(columns = {"rating": "num_of_ratings"}, inplace = True)
```

```python
num_ratings.head()
```

```python
books_with_ratings.head()
```

```python
final_ratings = books_with_ratings.merge(num_ratings, on = 'title')
```

```python
final_ratings.head()
```

```python
final_ratings.shape
```

```python
final_ratings = final_ratings[final_ratings['num_of_ratings'] >= 50]
```

```python
final_ratings.head()
```

```python
final_ratings.sample(10)
```

```python
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['user_id', 'title'], inplace = True)
```

```python
final_ratings.shape
```

```python
book_pivot = final_ratings.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace = True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
import pickle
pickle.dump(model, open('Artifacts/model.pkl', 'wb'))
pickle.dump(book_names, open('Artifacts/book_names.pkl', 'wb'))
pickle.dump(final_ratings, open('Artifacts/final_ratings.pkl', 'wb'))
pickle.dump(book_pivot, open('Artifacts/book_pivot.pkl', 'wb'))
```

```python
def recommend_books(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
book_name = "Exclusive"
recommend_books(book_name)
```
